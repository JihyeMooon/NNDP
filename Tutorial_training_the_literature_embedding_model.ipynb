```python
import pathlib
import sys
sys.path.append('lib')  
import Building_Literature_Embedding_Model as edg
```

```python
# Define hyper-parameters
context_size = 4
min_count = 3
min_size = 2
dimension = 128
num_sampled = 64
batch_size = 256 
epoch = 10
```

```python
root_path = '../data/preprocessed_data'
output='../data/tutorial_literature_embedding_model'  
  
vocab_dir = output + '/vocab/'
preprocessed_path = root_path + '/preprocessed/data.doc.txt'
model_path = output 
logs_dir = vocab_dir+'/logs'
gene2doc_dir = logs_dir+'/gene2doc' 

pathlib.Path(logs_dir).mkdir(parents=True, exist_ok=True)
pathlib.Path(gene2doc_dir).mkdir(parents=True, exist_ok=True) 
pathlib.Path(model_path).mkdir(parents=True, exist_ok=True)
```

```python
print("==== Generating Training Data for Literature Embedding Model")
eg=edg.building_embedding_model()
eg.setting(preprocessed_path, vocab_dir, logs_dir, gene2doc_dir)

print("==== Creating Vocabulary ===")
eg.creating_vocab(min_count = min_count, min_size = min_size)
```

```python
print("=== Checking If Data Generation Is Correct ===")
eg.checking_gene2doc_generation(context_size) # Fig. 2 in our visualization paper
```

```python
#print("=== Creating Training Data For Fig. 2 ===")
eg.creating_training_data_for_gene2doc(context_size)
 
#print("=== Starting Model Training For Fig. 2 ===")
eg.model_setting(dimension=dimension, num_sampled=num_sampled)
eg.starting_sorting(model_path)
eg.model_training(epoch=epoch, batch_size=batch_size, window_size = context_size)
```
